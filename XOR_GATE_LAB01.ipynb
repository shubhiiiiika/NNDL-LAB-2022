```python
import numpy as np # numpy lib makes working with array easy
from keras.models import Sequential 
from keras.layers.core import Dense #Dense respresenting number of neurons represnting the layer
```

```python
#traiing datasets for XOR GATE:
training_data= np.array([[0,0],[0,1],[1,0],[1,1]],"float32")
target_data = np.array([[0],[1],[1],[0]],"float32")
```

```python
model = Sequential()
model.add(Dense(16,input_dim=2,activation='relu')) #for i/p 16 hidden layer
model.add(Dense(1,activation='sigmoid')) #for o/p neural network layer
```

```python
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

```python
model.fit(training_data, target_data, epochs=500, verbose=2)
#It is neither np_epoch nor epoch, it is epochs
```

```python
model.predict(training_data).round() #don't use print
```
